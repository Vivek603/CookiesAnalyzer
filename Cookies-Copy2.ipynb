```python
combined_df = pd.DataFrame(columns=['Name', 'Value.page_transition', 'Value.title', 'Value.url', 'Value.client_id', 'Value.time_usec', 'domain', 'Time_Spent'])

# Initialize variables to track the previous row's values
prev_row = None
prev_time = None

# Iterate over each row in the DataFrame
for index, row in data.iterrows():
    print(index)

    # Check if the current row has the same values as the previous row
    if prev_row is not None and (row['Value.title']== prev_row['Value.title']):
        # Update the time spent
        time_spent =  prev_time - row['Value.time_usec']
#         print('this is time spent:', time_spent)
        combined_df.at[index - 1, 'Time_Spent'] = time_spent
    else:
        # Add a new row to the combined DataFrame
        combined_df = pd.concat([combined_df, pd.DataFrame(row).transpose()])

    # Update the previous row and time
    prev_row = row
    prev_time = row['Value.time_usec']


# Calculate time spent for the last record
combined_df.at[len(combined_df) - 1, 'Time_Spent'] = 0

combined_df.dropna(inplace=True)
combined_df.reset_index(inplace=True)
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
combined_df['Time_Spent'] = pd.to_timedelta(combined_df['Time_Spent'])

combined_df['Time_Spent_Seconds'] = combined_df['Time_Spent'].apply(lambda x: x.total_seconds())

kmeans = KMeans(n_clusters=4, random_state=42)
combined_df['Time_Spent_Cluster'] = kmeans.fit_predict(combined_df[['Time_Spent_Seconds']])

# Assign categories based on the clusters
category_map = {
    0: 'Low',
    1: 'Medium',
    2: 'High',
    3: 'Very High'
}
combined_df['Time_Spent_Category'] = combined_df['Time_Spent_Cluster'].map(category_map)

# # View the results
# print(combined_df[['Time_Spent_Seconds', 'Time_Spent_Category']])
```

```python
def log_transform(combined_df, col):
    for colname in col:
        if (combined_df[colname] == 0).any():
            # Add a small constant (e.g., 0.001) to handle zero values
            combined_df[colname + '_log'] = np.log(combined_df[colname] + 0.001)
        else:
            combined_df[colname + '_log'] = np.log(combined_df[colname])
    combined_df.info()
```

```python
log_transform(combined_df,['Time_Spent_Seconds'])
sns.distplot(combined_df['Time_Spent_Seconds_log'], axlabel='Time_Spent_Seconds');
```
